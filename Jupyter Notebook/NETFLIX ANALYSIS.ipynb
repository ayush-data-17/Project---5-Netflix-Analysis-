## **DATA PREPROCESSING AND CLEANING**

### **TASK TO DO :**
 1 . FIND THE NULL VALUES AND DUPLICATES 

 2 . FILL THE NULL VALUES OR DROP THEM 

3 . CREATE NEW COLUMNS/FEATURES THAT CAN HELP IN THE ANALYSIS 

 4 . CONVERT THE DATA TYPE OF THE COLLUMNS THAT ARE REQUIRED TO DO 

```python
## find the duplicates i the data 
df.duplicated().sum()
## no duplicates found in the data 
```

```python
## find the missing/null values in the data 
df.isna().sum()
missing_percent = round((df.isna().sum()/len(df))*100 , 2)
missing_percent
## approxematily 30% director names are missing from the data
## cast column has 9.37 missing values in the data 
## country column have 9.44% missing values in the data 
```

```python
## filling the missing values 
## in director column fill as "Director Not Found"
## in country column fill as "Unknown"
## in cast column fill as "Not Available"
df["director"] = df["director"].fillna("Director Not Found")
df["cast"] = df["cast"].fillna("Not Avaliable")
df["country"] = df["country"].fillna(df["country"].mode()[0])
df["duration"] = df["duration"].fillna(df["duration"].mode()[0])
df.isna().sum()
#df.dropna()
```

```python
## lets drop the cdescription column from the data there is no use of it in the entire analysis 
df.drop(columns = ["description"])
```

```python
## now lets exttract day .monthname , year , month from the column date_added and create new columns 
df["date_added"] = pd.to_datetime(df["date_added"] , errors = "coerce")
df["Month_Added"] = df["date_added"].dt.month
df["Month_Name"] = df["date_added"].dt.month_name()
df["Year"] = df["date_added"].dt.year
df["Day"] = df["date_added"].dt.day_name()
month_order = ["January" , 
               "Feburary" ,
               "March" , 
               "April" , 
               "May" , 
               "June" , 
               "July" , 
               "August" , 
               "September" , 
               "October" , 
               "November" , 
               "December" ]
df["Month_Name"] = pd.Categorical(df["Month_Name"] , categories = month_order , ordered = True)
```

```python
df['age_group'] = df['rating'].map({
    'TV-MA': 'Adults',
    'R': 'Adults',
    'PG-13': 'Teens',
    'TV-14': 'Teens',
    'TV-PG': 'Older Kids',
    'NR': 'Adults',
    'TV-G': 'Kids',
    'TV-Y': 'Kids',
    'TV-Y7': 'Older Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'NC-17': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'UR': 'Adults'
})
```

```python
df.dtypes
```

```python
## crating a new columns for genres 
genres = df["listed_in"].str.split(' ')

df['genre1'] = genres.str[0]
df['genre2'] = genres.str[1]  
df['genre3'] = genres.str[2]
```

### **2 . MOST CONTENT CREATING COUNTRIES**

```python
content_countries = df["country"].value_counts().head(10)
content_countries
```

```python
plt.figure(figsize =(10,6))
bars = plt.bar(content_countries.index, content_countries.values)
plt.xticks(rotation = -90)
plt.xlabel("Country")
plt.ylabel("Count")
plt.title("Most Content Creating Country")

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/ 2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha = "center" , va =  "bottom" , fontsize = 8 )

plt.show()
    
## UNITED KINGDOM , INDIA AND USA ARE THE MOSTCONTENT CREATING COUNTRY
```
